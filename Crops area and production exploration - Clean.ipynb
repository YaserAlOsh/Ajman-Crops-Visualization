## Data Cleaning

```python
ajman_pa_df.drop(['Service_AR','Service_EN','Category_AR','Category_EN'],axis=1,inplace=True)
ajman_pa_df.rename(columns={'Total Planted Area in Ajman (m2)':'area','Year':'year'},inplace=True)
ajman_pa_df
```

```python
#Make a copy just in case
clean_crop_df = crop_df.copy()
#Remove redundant columns
clean_crop_df.drop(['DATAFLOW','FREQ: Frequency of observation','OBS_STATUS: Observation status','UNIT_MULT: Unit multiplier','SOURCE_DETAIL: Source','OBS_COMMENT: Footnotes','DECIMALS: Decimals'],axis=1,inplace=True);
```

```python
clean_crop_df.rename(columns={"OBS_VALUE":"value","TIME_PERIOD: Time period":"year"},inplace=True)
```

```python
clean_crop_df = clean_crop_df.astype("string")
clean_crop_df['year'] = clean_crop_df['year'].astype(int)
clean_crop_df['value'] = clean_crop_df['value'].astype(float)
```

```python
#Removes the redundant codes, so: 'TIMEPERIOD: Time period' becomes 'Time period' only
clean_crop_df.rename(columns=lambda s : s[(s.index(':')+1 if s.find(':')>0 else 0):].lstrip().lower().replace(' ','_'),inplace=True)
```

```python
#Same as above, but does it to every cell of string type
clean_crop_df[['reference_area','unit_of_measure','crop_measure','crop_type','crop_name','crop_unit']]= clean_crop_df[['reference_area','unit_of_measure','crop_measure','crop_type','crop_name','crop_unit']].apply(lambda s: s.str.extract('(:)(\s*)(.*$)').iloc[:,2])
```

```python
def percentage_blocks(df, var,value):
    """
    Take as input a dataframe and variable, and return a Pandas series with
    approximate percentage values for filling out a waffle plot.
    """
    # compute base quotas
    df.index = df[var]
    percentages = 100 * df[value] / df[value].sum()
    counts = np.floor(percentages).astype(int) # integer part = minimum quota
    decimal = (percentages - counts).sort_values(ascending = False)
    # add in additional counts to reach 100
    rem = 100 - counts.sum()
    for cat in decimal.index[:rem]:
        counts[cat] += 1

    return counts.sort_values(ascending=False)
```

```python
waffle_counts = percentage_blocks(ajman_crops_area_2018, 'crop_type','value')
plt.figure(figsize=(9,9))
prev_count = 0
prev_cat = 0
# for each category,
for cat in range(waffle_counts.shape[0]):
    # get the block indices
    
    blocks = np.arange(prev_count, prev_count + waffle_counts[cat])
    
    # and put a block at each index's location
    x = blocks % 10 # use mod operation to get ones digit
    y = blocks // 10 # use floor division to get tens digit
    plt.bar(x = x, height = 0.8 , width = 0.8, bottom = y,color=colors_dict[waffle_counts.index[cat]])

    prev_count += waffle_counts[cat]
    prev_cat = cat

# aesthetic wrangling
plt.legend(waffle_counts.index, bbox_to_anchor = (1, 0.5), loc = 6,prop={'size': 13})
plt.axis('off')
plt.axis('square');
plt.title('Area Distribution by Crop Type in Ajman 2018',size=18);
```

```python
ajman_crops_prod = crops_prod_totals.query("reference_area=='Ajman'").reset_index(drop=True)
ajman_crops_prod = ajman_crops_prod.sort_values(by=['value'])
ajman_crops_prod.drop(index=ajman_crops_prod.query('crop_type=="Total"').index,inplace=True)
```

```python
prod_and_quantity_ajman = pd.merge(ajman_crops_prod[['year','crop_type','value']],ajman_crops_quantity[['year','crop_type','value']],on=['year','crop_type'],how='inner',suffixes=['','_y'])
prod_and_quantity_ajman['prod_per_quantity'] = prod_and_quantity_ajman.value / prod_and_quantity_ajman.value_y
prod_and_quantity_ajman.drop(columns=['value','value_y'],inplace=True)
years = len(prod_and_quantity_ajman.year.value_counts())
grouped = prod_and_quantity_ajman.groupby(['crop_type']).agg({'prod_per_quantity': 'sum'})
grouped.prod_per_quantity /= years
grouped
```

```python
plt.figure(figsize=(16,10))
barwidth=0.25
plt.style.use('default');

plt.bar(grouped.index,grouped.prod_per_quantity,width=barwidth,color=dark_green);

plt.xlabel('Crop Type',size=14)
plt.ylabel('AED Thousand per Metric Tonne',size=14);
plt.yticks(size=13,color=dark_green);
plt.gca().yaxis.get_major_ticks()[0].label1.set_visible(False)
#plt.gca().yaxis.set_visible(False)
plt.xticks(size=13,color=dark_green);
plt.title('Thousand AED per Metric Tonne',color='grey',size=14,**{'fontname':'Calibri'})
plt.suptitle('Average Ratio of Production Value per Quantity in Ajman (2013-2018)',size=20,**{'fontname':'Calibri'})
locs,labels = plt.yticks()
for i, v in enumerate(grouped.prod_per_quantity):
    plt.text(xlocs[i], v-0.5, str("%.1f" % (v)+"x"),ha="center",size=28,color='white',**segoefont)

#Aesthetics
SetAxisAesthetics(dark_green)
#plt.savefig('Average Ratio of Production Value per Quantity in Ajman (2013-2018).png', transparent=True);
```
